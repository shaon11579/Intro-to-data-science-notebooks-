### How seaborn conceptualizes data visualization

+ relationship plotting (relplot)
    + x and y are both numerical.  (scatter plot)
    + x and y can be geographical coordinates. (scatter plot)
    + x can be time. (line plot)

+ categorical plotting (catplot)
    + x is categorical; y is numerical.
    + data for each category.
    + summary statistics for each category.

+ distribution plotting (distplot)
    + how the data distributes/spreads in its entire range.
 

```python
import pandas
iris = pandas.read_csv('../Datasets/iris.csv')
tips = pandas.read_csv('../Datasets/tips.csv')
temp = pandas.read_csv('../Datasets/global_temp_monthly.csv', parse_dates=['Date'])
houses = pandas.read_csv('../Datasets/cali_housing.csv')
```

```python
from seaborn import relplot, catplot, displot
```

#### Relplot

```python
relplot(data=houses, 
        x='longitude', 
        y='latitude', 
        size='population',
        hue='median_house_value',
       height=8,
)
```

```python
houses.head()
```

```python
relplot(data=iris, x='SepalLength', y='SepalWidth')
```

```python
relplot(data=iris, x='SepalLength', y='SepalWidth', hue='Species')
```

```python
relplot(data=iris, x='SepalLength', y='SepalWidth', col='Species')
```

```python
relplot(data=iris, x='SepalLength', y='SepalWidth', hue='Species', col='Species')
```

```python
temp.head(9)
```

```python
relplot(data=temp, x='Date', y='Mean', col='Source', kind='line')
```

Line plots are used when there is an order between consecutive points on the x-axis.

### Categorical plots (catplot)

x - categorical

y - numerical

```python
catplot(data=iris, x='Species', y='PetalWidth')
```

```python
catplot(data=iris, x='Species', y='PetalWidth', kind='bar')
```

```python
catplot(data=iris, x='Species', y='PetalWidth', kind='box')
```

```python
catplot(data=iris, y='Species', x='PetalWidth', kind='violin')
```

```python
df = iris.sample(21)
```

```python
catplot(data=df, x='Species', y='PetalWidth', kind='violin')
```

```python
catplot(data=df, x='Species', y='PetalWidth', kind='box')
```

```python
catplot(data=df, x='Species', y='PetalWidth', kind='swarm')
```

#### Distribution plot (displot)

```python
displot(data=iris, x='PetalWidth', stat='percent', kind='hist', hue='Species', col='Species')
```

```python
displot(data=iris, x='PetalWidth', kind='kde', hue='Species', col='Species')
```

```python
displot(data=iris, x='PetalWidth', kind='kde', hue='Species')
```

```python
displot(data=iris, x='PetalWidth', y='PetalLength', kind='kde', hue='Species')
```

```python
displot(data=iris, x='PetalWidth', y='PetalLength', kind='kde', hue='Species', fill=True,rug=True)
```

```python
displot(data=iris, x='PetalWidth', y='PetalLength', kind='kde', hue='Species',rug=True)
```
