### Case Study 1: Relate Sepal Width and Petal Width to Petal Length

```python
from matplotlib import pyplot
from sklearn import tree
import pandas, seaborn
from dtreeviz.trees import *
from sklearn.tree import DecisionTreeRegressor
df = pandas.read_csv('../Datasets/iris50.csv')
print(df.sample(10))
print(df.var().round(2))
```

```python
X = df[['PetalWidth','SepalWidth']]
y = df['PetalLength']
model = DecisionTreeRegressor(max_depth=3)
model.fit(X,y)
print(model.score(X,y))
# viz = dtreeviz(model, X, y, feature_names=X.columns)
# viz.view()
fig = pyplot.figure(figsize=(20,15))
_=tree.plot_tree(model, feature_names=X.columns)
```
