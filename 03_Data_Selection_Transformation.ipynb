```python
def get_data(data_src):
    data = {}
    with open(data_src) as fp:
        header = fp.readline().strip()
        for line in fp:
            items = line.strip().split(',')
            species_name = items.pop()
            items = [ float(x) for x in items ]  
            if species_name in data:
                data[ species_name ].append(items)
            else:
                data[ species_name ] = [items]
    return header, data
```

# Data Transformation

Compute centroids for each species.

Create a new columns, p_setosa, which is the probability of data points belonging to the species setosa.

Create two new columns: p_virginica and p_versicolor.

```python
def get_centroid(species_data, header_indices):
    centroid = []
    for i in header_indices:
        sum_feature = 0
        for row in species_data:
            sum_feature += row[i]
        centroid.append(sum_feature / len(species_data))
    return centroid

header_indices = range(len(header.strip().split(',')) - 1)
centroids = {}
for species in data:
    centroids[species] = get_centroid(data[species], header_indices)
print(centroids)    
```

```python
from scipy.spatial.distance import cosine

def get_cosine_sim(centroid, data):
    similarity = []
    for species in data:
        for row in data[species]:
            cos_sim = 1 - cosine(centroid, row)
            similarity.append(cos_sim)
    return similarity

p_setosa = get_cosine_sim(centroids['setosa'], data)
p_versicolor = get_cosine_sim(centroids['versicolor'], data)
p_virginica = get_cosine_sim(centroids['virginica'], data)
```

```python
def add_column(data, spe_prob):
    i = 0
    for species in data:
        for row in data[species]:
            row.append(spe_prob[i])
            i += 1
    return data

data = add_column(data, p_setosa)
data = add_column(data, p_versicolor)
data = add_column(data, p_virginica)
```
